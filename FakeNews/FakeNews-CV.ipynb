```python
#importing standard libraries
import pandas as pd
import numpy as np
```

```python
#reading dataset
df=pd.read_csv("news.csv", index_col=None)
```

```python
df.head()
```

```python
# removing unwanted column
dataset = df.drop('Unnamed: 0', axis=1)
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
#Splitting dataset and assing to independent variable (X) and dependent variable (y)
X=dataset['text']
y=dataset['label']
```

```python
# Splitting data for train and test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=53)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# importing vector libraries
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
```

```python
# converting independent variable into vector
count_vectorizer = CountVectorizer(stop_words='english')  #stop_words like is, was, the, a kind of supporting words will be ignored 
count_train = count_vectorizer.fit_transform(X_train) # transform the training data to vector db
count_test = count_vectorizer.transform(X_test) # transform the test data based on training words to vector db
```

```python
len(count_vectorizer.get_feature_names_out())
```

```python
count_vectorizer.get_feature_names_out()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mn_clf = MultinomialNB()
```

```python
mn_clf.fit(count_train, y_train)
```

```python
y_pred = mn_clf.predict(count_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred, labels=['FAKE', 'REAL'])
print(cm)
```

```python
clf_report = classification_report(y_test, y_pred)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print (accuracy_score(y_test, y_pred))
```
